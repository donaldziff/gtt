```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import altair as alt

```

```python
df = pd.read_csv('clean-results.csv')
```

```python
df = df[df['epoch']<100]
```

```python
alt.Chart(df[df['model'].isin(['BERT-Base'])]).mark_line().encode(
    x='epoch',
    y='mAvg:F1',
    color='model',
).properties(
    width=800,
    height=300,
    title='Figure 1'
)
```

```python
alt.Chart(df[df['model'].isin(['BERT-Base', 'BERT-Medium'])]).mark_line().encode(
    x='epoch',
    y='mAvg:F1',
    color='model',
).properties(
    width=800,
    height=300,
    title='Figure 2'
)
```
